```
def cat_alumno(df):
    val_temp = 0
    cont_temp = 0
    if  df['201901'] == 27 :
        val_temp = val_temp + df['201901'] * 0.65
        cont_temp = cont_temp + 1
    elif df['201901'] == 23 :
        val_temp = val_temp + df['201901'] * 0.25
        cont_temp = cont_temp + 1
    elif df['201901'] == 18 :
        val_temp = val_temp + df['201901'] * 0.1
        cont_temp = cont_temp + 1

    if df['201902'] == 27 :
        val_temp = val_temp + df['201902'] * 0.65
        cont_temp = cont_temp + 1
    elif df['201902'] == 23 :
        val_temp = val_temp + df['201902'] * 0.25
        cont_temp = cont_temp + 1
    elif df['201902'] == 18 :
        val_temp = val_temp + df['201902'] * 0.1
        cont_temp = cont_temp + 1

    if df['202001'] == 27 :
        val_temp = val_temp + df['202001'] * 0.65
        cont_temp = cont_temp + 1
    elif df['202001'] == 23 :
        val_temp = val_temp + df['202001'] * 0.25
        cont_temp = cont_temp + 1
    elif df['202001'] == 18 :
        val_temp = val_temp + df['202001'] * 0.1
        cont_temp = cont_temp + 1
    
    if df['202002'] == 27 :
        val_temp = val_temp + df['202002'] * 0.65
        cont_temp = cont_temp + 1
    elif df['202002'] == 23 :
        val_temp = val_temp + df['202002'] * 0.25
        cont_temp = cont_temp + 1
    elif df['202002'] == 18 :
        val_temp = val_temp + df['202002'] * 0.1
        cont_temp = cont_temp + 1

    val_temp = val_temp / cont_temp
    return val_temp
```

```
df_alumno['CategoriaAlumno'] = df_alumno.apply(cat_alumno, axis=1)
```

```
def type_Alumno(df):
    # SE DEBE AFINAR AQUELLAS LISTAS QUE TIENEN AL MENOS UNA CELDA VACIA
    '''
    Si un alumno tiene en los 4 semestre 18 cred como max tiene 1
    Si un alumno tiene en los 4 semestre 23 cred como max tiene 4
    Si un alumno tiene en los 4 semestre 27 cred como max tiene 5

    Si un alumno tiene en los últimos periodos 18 cred max y en uno mayor a este, tiene 2
    Si un alumno tiene en en sus periodos 23 y 27 y en el último 18  tiene 3
    Si un alumno tiene en el último periodo 27 cred y en sus últimos periodos variabilidad tiene 3.5
    Si un alumno termina en su último ciclo con 27 tiene 4
    Si un alumno termina en su último ciclo con 27 y en unos de sus últimos 4 periodos 23 tiene 4.5
    '''
    lastPerio=df[['202002']].values
    primerPerio=df[['201901']].values
    tolist = df[['201901','201902','202001','202002']].values.tolist()
    counttolist = {x:tolist.count(x) for x in tolist}
    
    if sum(df[['201901','201902','202001','202002']])/4 == 18 : # usar la funcion promedio
        return 1
    elif sum(df[['201901','201902','202001','202002']])/4 == 23:
        return 4
    elif sum(df[['201901','201902','202001','202002']])/4 == 27:
        return 5
    elif ( 18 in counttolist and counttolist[18] >= 2 and 27 in counttolist and counttolist[27] >= 1 ) or ( 18 in counttolist and counttolist[18] >= 2 and 23 in counttolist and counttolist[23] >= 1 ):
        return 2
    elif ( lastPerio == 18 and counttolist[18] == 1 and 23 in counttolist and counttolist[23] >= 1 ) or ( lastPerio == 18 and counttolist[18] == 1 and 27 in counttolist and counttolist[27] >= 1 ):
        return 3
    #elif ( primerPerio == 18 and counttolist[18] == 1 and 23 in counttolist and counttolist[23] >= 1 ) or ( lastPerio == 18 and counttolist[18] == 1 and 27 in counttolist and counttolist[27] >= 1 ):
    #    return 3
    elif lastPerio == 27 and 18 in counttolist and counttolist[18] >= 1:
        return 3.5
    elif lastPerio == 27 and 23 in counttolist and counttolist[23] >= 3:
        return 4.5
    elif lastPerio == 27:
        return 4
    else:
        return tolist   
    
```
